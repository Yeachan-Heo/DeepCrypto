```python
df = load_bitmex_data("/home/ych/Storage/bitmex/bitmex.db", "1H", SYMBOL)
```

```python
def seungwook_rsi_strategy(cfg, df):
    df = df.backtest.add_defaults()

    df["RSI"] = RSI(df["close"], cfg["RSI"])

    df["volume_filter"] = df["volume"] / df["volume"].rolling(50).mean() > cfg["VOL"]

    df["enter_long"] = (df["RSI"]/df["RSI"].rolling(50).mean() > 1 + cfg["RSI_T"]) & df["volume_filter"]
    df["enter_short"] = (df["RSI"]/df["RSI"].rolling(50).mean() < 1 - cfg["RSI_T"]) & df["volume_filter"]

    df["time_cut"] = cfg["TC"]
    df["take_profit"] = cfg["TP"]
    df["stop_loss"] = cfg["SL"]
    df["trade_cost"] = 0.0005
    df["bet"] = np.clip(0.01 / df["close"].pct_change().rolling(50).std(), 0.1, 1)
    df["low_first"] = 1
    return df
```

```python
config = {"RSI": 24, "RSI_T": 0.01, "SL": 0.02, "TC": 200, "TP": 0.2, "VOL": 6}
```

```python
output, order = seungwook_rsi_strategy(config, df).backtest.run()
output = output.resample("1D").last()

print(all_metrics(output,order))
qs.reports.full(output["portfolio_value"])
```

```python
from ray import tune
import ray

def optimizer(config, df):
    from deepcrypto.backtest import BacktestAccessor
    portfolio_df, order_df = seungwook_rsi_strategy(config, df).backtest.run()
    tune.report(**all_metrics(portfolio_df, order_df))

def optimize(df):
    result = tune.run(
        tune.with_parameters(optimizer, df=df),
        config = {
            "RSI" : tune.grid_search([12 * i for i in range(1, 8, 2)]), 
            "VOL" : tune.grid_search(list(range(3, 8))), 
            "RSI_T" : tune.grid_search([0, 0.01, 0.02]), 
            "TC" : tune.grid_search([24 * i for i in range(1, 8, 2)]),
            "TP" : tune.grid_search([0.05, 0.1, 0.2]), 
            "SL" : tune.grid_search([0.02, 0.05])
        },
        local_dir="./ray_results",
    )
    return result
```
